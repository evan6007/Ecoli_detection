```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python

img = cv2.imread('/content/Ecoli_detection/Ecoli (1).jpg', cv2.IMREAD_COLOR) #把照片路徑放進去
img_orig = img.copy()
img_orig = cv2.cvtColor(img_orig, cv2.COLOR_BGR2RGB)
plt.rcParams["figure.figsize"] = (16,9)
plt.imshow(img_orig)
```

```python
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
low_hsv = np.array([0,43,46])
high_hsv = np.array([10,255,255])
mask = cv2.inRange(hsv,lowerb=low_hsv,upperb=high_hsv)
plt.rcParams["figure.figsize"] = (16,9)
plt.imshow(mask)
img=mask
```

```python
img = cv2.GaussianBlur(img, (21,21), cv2.BORDER_DEFAULT)
plt.rcParams["figure.figsize"] = (16,9)
plt.imshow(img,cmap='gray')


all_circle =cv2.HoughCircles(img, cv2.HOUGH_GRADIENT, 0.8, 30,
              param1=50,
              param2=25,
              minRadius=0,
              maxRadius=70)
all_circle_rounded = np.uint16(np.around(all_circle))
```

```python
img = cv2.imread('/content/Ecoli_detection/Ecoli (44).jpg', cv2.IMREAD_COLOR) #把照片路徑放進去
img_orig = img.copy()
img_orig = cv2.cvtColor(img_orig, cv2.COLOR_BGR2RGB)

count = 1
for i in all_circle_rounded[0,:]:
    cv2.circle(img_orig,(i[0],i[1]),i[2],(50,200,200),5)
    cv2.circle(img_orig,(i[0],i[1]),2,(255,0,0),3)
    count = count +1



plt.rcParams["figure.figsize"]=(16,9)
plt.imshow(img_orig)
print("有" + str(all_circle_rounded.shape[1])+'顆Ecoli')
```
